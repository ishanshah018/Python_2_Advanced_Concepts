```python
# QB-72

# The following dictionary shows how five people follow each other on Instagram:
#  instagram = {'person1': [0,1,1,0,1], 'person2': [0,0,1,0,1], 'person3': [1,1,0,1,1],'person4': [1,1,1,0,0], 'person5': [1,1,0,0,0]}
#  E.g., the list for person1 has the value on index 2 as 1 which means person1 follows person3 and a directed edge should be
# added from person1 to person3.
#  Using networkx library, create a directed graph

import networkx as nx
import matplotlib.pyplot as plt

instagram = {
    'person1': [0, 1, 1, 0, 1],
    'person2': [0, 0, 1, 0, 1],
    'person3': [1, 1, 0, 1, 1],
    'person4': [1, 1, 1, 0, 0],
    'person5': [1, 1, 0, 0, 0]
}

people = list(instagram.keys())

G = nx.DiGraph()

# Add edges based on follow (1 means follows)
for i, person in enumerate(people):
    for j, follow in enumerate(instagram[person]):
        if follow == 1:
            G.add_edge(person, people[j])

# Draw the directed graph
nx.draw(G, with_labels=True, arrows=True)
plt.show()
```

```python
# QB-85

# You have been hired as a network analyst by a company to analyze the social network of their employees. The company has
# provided you with the following data:
#  There are 5 employees in the company, each identified by a unique ID from 1 to 5.
#  The following relationships exist between the employees:
#  1. Employee 1 is friends with Employee 2 and Employee 3.
#  2. Employee 2 is friends with Employee 4.
#  3. Employee 3 is friends with Employee 5.
#  Your task is to create a NetworkX graph representing this social network and display it.

import networkx as nx
import matplotlib.pyplot as plt

g=nx.DiGraph()
employees=['Employee 1','Employee 2','Employee 3','Employee 4','Employee 5']

friendship=[('Employee 1','Employee 2'),
            ('Employee 1','Employee 3'),
            ('Employee 2','Employee 4'),
            ('Employee 3','Employee 5')]

g.add_nodes_from(employees)

for u,v in friendship:
    g.add_edge(u,v)

nx.draw(g,with_labels=True)
plt.show()
```

```python
# QB-86

# Consider the following numpy arrays:
# Time=np.arange(12)
# income=np.array([5,9,6,6,10,7,6,4,4,5,6,4])
# expense=np.array([6,6,8,3,6,9,7,8,6,6,4,8])
# Use Time array for X-axis and create two separate lines in the same graph with income & expense on Y-axis. Give Appropriate
# labels. Create an area fill graph between the two lines in such a way that where income is more than expense, are filled with
# Green and areas where expense is more than income are filled with red. 
import numpy as np

time=np.arange(12)
income=np.array([5,9,6,6,10,7,6,4,4,5,6,4])
expense=np.array([6,6,8,3,6,9,7,8,6,6,4,8])


plt.plot(time,income,color='green',label='Income')
plt.plot(time,expense,color='red',label='Expense')

plt.fill_between(time,income,expense,where=(income>expense),color='green',label='Positive')
plt.fill_between(time,income,expense,where=(income<expense),color='red',label='Negative')

plt.xlabel('time')
plt.ylabel('income/expense')
plt.legend()
plt.show()
```
